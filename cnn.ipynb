```python
import imageio
import matplotlib.pyplot as plt
%matplotlib inline

# Read image
im = imageio.imread("/home/marriane/Pictures/BKE_5459.jpg")
print("Shape of the image tensor: {}".format(im.shape))
plt.imshow(im)
```

```python
#our image has a height of 6000 and a width of 4000 and both the three colors
#we can easily slice out the colors
# Red channel
im_red = im[:,:,0]
print(im_red.shape)
plt.imshow(im_red, cmap='gray')
```

```python
# Blue channel
im_blue = im[:,:,2]
print(im_red.shape)
plt.imshow(im_blue, cmap='gray')
```

```python
# green channel
im_green = im[:,:,1]
print(im_green.shape)
plt.imshow(im_green, cmap='gray')
```
